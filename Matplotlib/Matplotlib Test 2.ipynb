```python
data = pd.read_csv("countries.csv")  # Get all the data from this csv file, and save it as a data frane.

# Get only the rows where the country column maches "United States". In other words, only the US data.
US = data[data.country == "United States"]

# Get only the rows where the country column maches "China". In other words, only the China data.
China = data[data.country == "China"]
```

```python
# Plot the growth of the US and Chinese population.
plt.plot(US.year, US.population / 10**6)
plt.plot(China.year, China.population / 1000000)  # 1000000 = 10**6

# Label the x and y side of the graph accordingly.
plt.xlabel("Years")
plt.ylabel("Population")

plt.legend(["US population", "China population"])  # This will let us know what each line represents.
plt.show()
```

```python
print(US.population / US.population.iloc[0] * 100, "\n")  # The US population as a percent.
print(China.population / China.population.iloc[0] * 100)  # The China population as a percent.
```

```python
# Plot the growth percent of the US and Chinese population.
plt.plot(US.year, US.population / US.population.iloc[0] * 100)
plt.plot(China.year, China.population / China.population.iloc[0] * 100)

# Label the x and y side of the graph accordingly.
plt.xlabel("Years")
plt.ylabel("Population %")

plt.legend(["US population %", "China population %"])  # This will let us know what each line represents.
plt.show()
```

```python
# Plot a random line. We can add a label here so we don't have to for legend. We can also add a color.
# linewidth=5, it obviously the width. marker="." is the three dots on the line, each representing the numbers we're ploting.
# markersize=15 is obviously how large the dots are. 
# We could of used "g-." for the color, (g = green), line, (- is a full-line if it comes before the marker) and marker, (. The dots).
# But this way is more neat and understandable.
plt.plot([1, 2, 3], [4, 5, 6], label="Green Line", color="green", linewidth=5, marker=".", markersize=15)

# Change the tick size. The numbers to the left and bottem.
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

# Label the Horizontal and vertical side of the graph accordingly. Also add a font and size for it.
plt.xlabel("Horizontal", fontdict={"fontname": "Comic Sans MS", "fontsize": 10})
plt.ylabel("vertical", fontdict={"fontname": "Comic Sans MS", "fontsize": 10})

plt.title("Graph 5", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})  # Add a title, and give it a font, and font size.
plt.legend()  # Since we adding a label when plotting, we don't need to mention it. We still need legend to tell we want to use it.  
plt.show()
```

```python
apples = np.arange(0, 4.5, 0.5)  # numpy.arange(start, stop, step, dtype) Create am array of [ 1,  3,  5,  7,  9, 11]
```

```python
plt.figure(figsize=(5, 3), dpi=100)  # change the figure size to (5, 3). Then change the pixels per inch, (dpi=), to 200.

# Plot two lines to show how many apples we have and how many is too many. This is dones by cutting the line into two.
plt.plot(apples[:6], apples[:6]**2, "r", label="Apples")  # The apples in limits.
plt.plot(apples[5:], apples[5:]**2, "r--", label="Too many Apples")  # The apples over the limit, or too many.

# Label the Horizontal, (days), and vertical, (count), side of the graph accordingly. Also add a font and size for it.
plt.xlabel("Days", fontdict={"fontname": "Comic Sans MS", "fontsize": 10})
plt.ylabel("Count", fontdict={"fontname": "Comic Sans MS", "fontsize": 10})

# Give the graph a name a font.
plt.title("APPLE GRAPH", fontdict={"fontname": "Comic Sans MS", "fontsize": 20})

# Change the tick size. The numbers to the left and bottem.
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])
plt.legend()  # Since we adding a label when plotting, we don't need to mention it. We still need legend to tell we want to use it.

plt.savefig("graph_test1.png")  # Save the graph as "graph_test1.png"

plt.show()
```

```python
# I found all of this on matplotlib.org in the documentation. 

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis)

plt.show()
```

```python
plt.rcParams['axes.facecolor'] = '#123'  # Make the inside dark blue rather then purple.

fifa = pd.read_csv("fifa_data.csv.txt", delimiter=",")  # It's time for some new data!
fifa
```

```python
left_ = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]  # How many players prefer using their left foot, which is 4211.
right_ = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]  # How many players prefer using their right foot, which is 13948.
print(left_)
print(right_)

# Use a pie graph to see how many players prefer using their left foot, vs their right. (As a percent.)
plt.pie([left_, right_], labels=["Left", "Right"], colors=["#115", "#511"])

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]  # Turn all the ibs, or pounds, into ints. Removing the str.
print(fifa.Weight)  # Be sure it worked.
```

```python
# We'll need six variables...
light = fifa.loc[fifa.Weight < 125].count()[0]  # < 125.
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]  # 125 to 150.
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]  # 150 to 175.
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]  # 175 to 200
heavy = fifa[(fifa.Weight >= 200)].count()[0] # >= 200
labels = ['under 125', '125-150', '150-175', '175-200', 'over 200']
```

```python
explode_ = [0.5, 0.2, 0, 0, 0.4]  # The list we'll want to use for the explode function in pie.
plt.style.use("ggplot")  # Change the style of the pie graph.

# Use the variables from above to create and label a pie graph. Use autopct='%.2f %%' to show the percent on each piece of pie.
# pctdistance is how far from the center those percent labels will be. explode cuts the pie so numbers don't have to mix.
plt.pie([light, light_medium, medium, medium_heavy, heavy], labels=labels, autopct='%.2f %%', pctdistance=0.7, explode=explode_)

plt.show()
```

```python
barcelona = fifa.loc[fifa["Club"] == "FC Barcelona"]["Overall"]  # Get FC Barcelona's overall score.
madrid = fifa.loc[fifa["Club"] == "Real Madrid"]["Overall"]  # Get Real Madrid's overall score.
NE = fifa.loc[fifa["Club"] == "New England Revolution"]["Overall"]  # Get New England Revolution's overall score.
labels_ = ['FC Barcelona', "Real Madrid", "NE Revolution"]
print(barcelona)
print(madrid)
print(NE)
```

```python
plt.boxplot([barcelona, madrid, NE], labels=labels_)  # Using the variables from above, create a box plot, with labels.
plt.show()
```

```python
# save boxes as a dict, with all the objects inside. We can use these to plot with in a for loop.
# This will still let us be able to show the graph. If patch_artist is True, then we can use color and fill in the boxes.
# medianprops={"linewidth": 2} this is how wide the line inside the box will be.
boxes = plt.boxplot([barcelona, madrid, NE], labels=labels_, patch_artist=True, medianprops={"linewidth": 2})  

# Look though boxes["boxes"]
for box in boxes["boxes"]:
# [<matplotlib.patches.PathPatch object at 0x128D2DA8>, "boxes"
#  <matplotlib.patches.PathPatch object at 0x128D5928>, 
#  <matplotlib.patches.PathPatch object at 0x12D324A8>]
    box.set(color='#116')  # Change the color of them all to blue
        
plt.show()
```
