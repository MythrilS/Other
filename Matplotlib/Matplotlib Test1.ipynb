```python
fifa = pd.read_csv("fifa_data.csv.txt", delimiter=",")
fifa
```

```python
b = [40, 50, 60, 70, 80, 90, 100]
plt.hist(fifa.Overall, bins=b)

plt.show()
```

```python
left_ = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right_ = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
print(left_)
print(right_)

plt.pie([left_, right_], labels=["Left", "Right"], autopct="%.2f %%", colors=['r', 'g'])

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x)==str else x for x in fifa.Weight]
```

```python
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
labels = ['under 125', '125-150', '150-175', '175-200', 'over 200']
```

```python
explode_ = [0.4, 0.2, 0.2, 0.2, 0.4]
plt.style.use("ggplot")
plt.pie([light, light_medium, medium, medium_heavy, heavy], labels=labels, autopct='%.2f %%', pctdistance=0.7, explode=explode_)

plt.show()
```

```python
barcelona = fifa.loc[fifa["Club"] == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa["Club"] == "Real Madrid"]["Overall"]
NE = fifa.loc[fifa["Club"] == "New England Revolution"]["Overall"]
labels_ = ['FC Barcelona', "Real Madrid", "New England Revolution"]
```

```python
boxes = plt.boxplot([barcelona, madrid, NE], labels=labels_, patch_artist=True, medianprops={"linewidth": 2})

for box in boxes["boxes"]:
    box.set(color="#e0e0e0", linewidth=2)

plt.show()
```
