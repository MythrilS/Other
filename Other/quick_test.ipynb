```python
# Set this up for later.
years = range(2009, 2020)
size = 0.3


def divide_pie(arr, number):
    array_ = np.array([])
    sum_ = sum(arr)
    
    for i in arr:
        x = (sum_/i)
        x = 100/x
        x = number/(50/i)
        array_ = np.append(array_, x)
        
    return np.array(array_, dtype=float)


# Create a donut chart for all the following.
for column in ["User Rating", "Reviews", "Price"]:
    
    array_ = np.array([])  # Reset the array.
    fiction_array = np.array([0, 0])

    # All the years as integers, one at a time, from 2009 to 2019.
    for yr in years:
        
        # Get a data frame for all the books in the specific year.
        year = df[df["Year"] == yr]

        # Get the sum of everthing in the specific column, except the year.
        column_sum = year[column].sum()
        
        fiction_count = list(year.Genre.value_counts())
        
        fiction_ = divide_pie(fiction_count, column_sum)
#         print(fiction_count, column_sum, fiction_)  # After this, I know math will be needed...
        
#         fiction_array = np.vstack([(year[year[column]])])
        fiction_array = np.vstack([fiction_array, fiction_])
#         print(fiction_count, column_sum)

        # Append the column sum to the array, for the donut chart later.
        array_ = np.append(array_, column_sum)

    fiction_array = np.delete(fiction_array, 0, 0)
#     print(fiction_array, array_)

    # vvv Code from matplotlib.org on how to make a donut chart. vvv
    fig, ax = plt.subplots(figsize=(10, 7), subplot_kw=dict(aspect="equal"))

    theme = plt.get_cmap('terrain')  # 'BuGn_r'
    ax.set_prop_cycle("color", [theme(1. * i / len(array_))
                                 for i in range(len(array_))])

    wedges, texts, temp = ax.pie(array_, wedgeprops=dict(width=size, edgecolor='w'), startangle=-40, autopct="%.1f%%", pctdistance=0.4, labels=years, radius=1)
    ax.pie(fiction_array.flatten(), wedgeprops=dict(width=size, edgecolor='w'), pctdistance=0.2, radius=1-size)
    [ _.set_fontsize(10) for _ in texts ]

    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    kw = dict(arrowprops=dict(arrowstyle="-"),
              bbox=bbox_props, zorder=0, va="center")

    ax.set_title(f"Annual {column} Percentage.")
    ax.set(aspect="equal", title='Pie plot with `ax.pie`')
    plt.show()
    print(array_, fiction_array.flatten())
    # ^^^ https://matplotlib.org/gallery/pie_and_polar_charts/pie_and_donut_labels.html?highlight=donut ^^^
```
